```python
class Node:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.feature = None
        self.threshold = None
        self.data = data
        self.m = 0
```

```python
def best_params_r(data):
    ftr = data.columns[0]
    trsld = data[ftr].unique()[0]
    err = data[data[ftr] <= trsld].iloc[:, -1].std() + data[data[ftr] > trsld].iloc[:, -1].std()
    for f in data.iloc[:, :-1].columns:
        for t in data[f].unique():
            e = data[data[f] <= t].iloc[:, -1].std() + data[data[f]  > t].iloc[:, -1].std()
            if e < err:
                err = e
                ftr = f
                trsld = t
    return ftr, trsld

def best_params_c(data):
    ftr = data.columns[0]
    trsld = data[ftr].unique()[0]
    d = data[data[ftr] <= trsld]
    m = d.iloc[:, -1].mode()[0]
    per = d[d.iloc[:, -1] == m].shape[0] / d.shape[0]
    for f in data.iloc[:, :-1].columns:
        for t in data[f].unique():
            d = data[data[f] <= t]
            m = d.iloc[:, -1].mode()[0]
            p = d[d.iloc[:, -1] == m].shape[0] / d.shape[0]
            if p > per:
                per = p
                ftr = f
                trsld = t
    return ftr, trsld
```

```python
def Rec_Tree(data, n, type, len):
    node = Node(data)
    print(data.shape[0])
    if data.shape[0] <= n or data.shape[0] == len:
        if type == 'r':
            node.m = node.data.iloc[:, -1].mean()
        if type == 'c':
            node.m = node.data.iloc[:, -1].mode()[0]
        return node
    else:
        if type == 'r':
            f, t = best_params_r(data)
            node.feature = f
            node.threshold = t
            node.left = Rec_Tree(data[data[f] <= t], n, 'r', data.shape[0])
            node.right = Rec_Tree(data[data[f] > t], n, 'r', data.shape[0])
        if type == 'c':
            f, t = best_params_c(data)
            node.feature = f
            node.threshold = t
            if data[data[f] <= t].shape[0] != 0:
                node.left = Rec_Tree(data[data[f] <= t], n, 'c', data.shape[0])
            else:
                return node
            if data[data[f] > t].shape[0] != 0:
                node.right = Rec_Tree(data[data[f] > t], n, 'c', data.shape[0])
            else:
                return node
        return node
```

```python
class Tree:
    def __init__(self):
        pass
    def fit_r(self, X, y, n=100):
        data = pd.concat([pd.DataFrame(X).reset_index(drop=True), pd.DataFrame(y).reset_index(drop=True)], axis=1)
        self.Tree_r = Rec_Tree(data, n, 'r', data.shape[0] + 1)
    def fit_c(self, X, y, n=100):
        data = pd.concat([pd.DataFrame(X).reset_index(drop=True), pd.DataFrame(y).reset_index(drop=True)], axis=1)
        self.Tree_c = Rec_Tree(data, n, 'c', data.shape[0] + 1)
    def predict_r(self, X):
        y_pred = []
        X = pd.DataFrame(X)
        X = X.reset_index(drop=True)
        for i in range(X.shape[0]):
            cur_node = self.Tree_r
            while cur_node.left != None or cur_node.right != None:
                if X[cur_node.feature][i] <= cur_node.threshold:
                    cur_node = cur_node.left
                else:
                    cur_node = cur_node.right
            y_pred.append(cur_node.m)
        return np.array(y_pred)
    def predict_c(self, X):
        y_pred = []
        X = pd.DataFrame(X)
        X = X.reset_index()
        for i in range(X.shape[0]):
            cur_node = self.Tree_c
            while cur_node.left != None or cur_node.right != None:
                if X[cur_node.feature][i] <= cur_node.threshold:
                    cur_node = cur_node.left
                else:
                    cur_node = cur_node.right
            y_pred.append(cur_node.m)
        return np.array(y_pred)
```
