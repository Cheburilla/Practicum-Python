5. С помощью библиотек matplotlib, seaborn и pandas построить 5 диаграмм

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df["Pistol_Usage"], binwidth=0.8)
plt.savefig('1.png')
```

```python
plt.boxplot(x=df['Hours_without_reallife'])
plt.xticks([1], ['Hours'])
plt.savefig('2.png')
```

```python
df['Difficulty'].value_counts()
plt.pie(df['Difficulty'].value_counts().values, labels=df['Difficulty'].value_counts().index)
plt.savefig('3.png')
```

```python
corr = df.corr()
sns.set(rc = {'figure.figsize':(40,40)})
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot_kws={"size":10})
plt.savefig('4.png')
```

```python
f = plt.figure(figsize=(10,4))
def mapper(value):
  if value > 100:
    return "Pro" 
  elif value > 40: 
    return "Normal"
  else: return "Noob"
df["Troll_Skills"] = df["Average_Friendly_Fire"].map(mapper)
sns.countplot(x = df["Difficulty"], hue=df["Troll_Skills"], color="magenta")
plt.savefig('5.png')
```

7. Выбрать произвольный числовой признак, с помощью среза данных выбрать не более 200 значений соответствующего признака. Проверить полученную выборку с помощью normaltest на нормальность распределения.

```python
import scipy as sp
import numpy as np

a = np.array(df['Molotov_Kills'][:200])
print(sp.stats.normaltest(a))
```

8. Выполнить one-hot кодирование всех категориальных признаков. 



```python
s = pd.Series(df.Troll_Skills)
s = pd.get_dummies(s)
s
```

```python
f = pd.Series(df.Difficulty)
f = pd.get_dummies(f)
f
```
